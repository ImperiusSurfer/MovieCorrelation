```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'/Users/waqaaspatel/Documents/Course/Python/movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# data types 
df.dtypes
```

```python
# change data type of columns

df['budget'].round().astype('Int64')
df['budget']=df['budget'].astype('Int64')
```

```python
df.head()
```

```python
# create correct year column


df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
df.head()
```

```python
 df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# drop any null values

df = df.dropna(how='any', axis=0, inplace=True)
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df ['gross'])

plt.title('Budget vs Gross')

plt.xlabel ('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```

```python
df.head
```

```python
# Plot budget vs gross earnings seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'cyan'})

plt.show()
```

```python
# let's start looking at the correlaations

# pearson, kendall, spearmen

df.corr(method='pearson', numeric_only=True)
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for numeric features')

plt.xlabel ('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

print(df_numerized.head())

```

```python
correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for numeric features')

plt.xlabel ('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs)> 0.5]

high_correlation
```
